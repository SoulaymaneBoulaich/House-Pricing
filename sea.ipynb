```python
import pandas as pd

data = pd.read_csv('train.csv')

data.head(10)
```

```python
data.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.histplot(data['SalePrice'],kde=True)
plt.title('Distribution of SalePrices')
plt.xlabel('SalePrice')
plt.ylabel('Frequency')
plt.show()
```

```python
#Select a certain columns to print them
selected_data = data.loc[:,['SalePrice', 'OverallQual', 'GrLivArea', 'GarageCars', 'TotalBsmtSF', 'YearBuilt']]

#Calculate the correlations
mat_corr = selected_data.corr()
mat_corr

#Visualize the correlation heat map
plt.figure(figsize=(10, 6)) #plt.figure(figsize=(width,height))
sns.heatmap(mat_corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Selected Features')
plt.show()
```

```python
# Scatter plot for OverallQual vs SalePrice
plt.figure(figsize=(10, 6))
sns.scatterplot(x='OverallQual', y='SalePrice', data=data)
plt.title('OverallQual vs SalePrice')
plt.xlabel('OverallQual')
plt.ylabel('SalePrice')
plt.show()

# Scatter plot for GrLivArea vs SalePrice
plt.figure(figsize=(10, 6))
sns.scatterplot(x='GrLivArea', y='SalePrice', data=data)
plt.title('GrLivArea vs SalePrice')
plt.xlabel('GrLivArea')
plt.ylabel('SalePrice')
plt.show()
```

```python
# Box plot for OverallQual
plt.figure(figsize=(13, 6))
sns.boxplot(x='OverallQual', y='SalePrice', data=data)
plt.title('SalePrice by OverallQual')
plt.xlabel('OverallQual')
plt.ylabel('SalePrice')
plt.show()

# Box plot for YearBuilt
plt.figure(figsize=(13, 6))
sns.boxplot(x='YearBuilt', y='SalePrice', data=data)
plt.title('SalePrice by YearBuilt')
plt.xlabel('YearBuilt')
plt.ylabel('SalePrice')
plt.xticks(rotation=90)
plt.show()
```
